## Import Package

```python
import itertools
from random import sample
import numpy as np
```

## Deck Generator

```python
def deck_generator(M=4, N=13):
    lst_suits = []
    lst_cards = []
    for i in range(1, M+1):
        lst_suits.append('M'+str(i))
    for i in range(1, N+1):
        lst_cards.append('N'+str(i))
    somelists = [lst_suits, lst_cards]
    res = []
    for element in itertools.product(*somelists):
        res.append(element)
    return res
```

$\mathbb{E}[X] = \mathbb{\Sigma}_{x\in A} X_i \times P(X_i)$

$P(X_i) = 12 \div 51$

$X(a, b)= 
\begin{cases}
    1,& \text{if } suit(a) = suit(b)\\
    0,              & \text{otherwise}
\end{cases}
$

$ 51 \times (12 \div 51) = 12$

## Question 1 & 2

```python
lst = deck_generator(M=2, N=26)
NumberOfChoice = 100000
point = []
for i in range(0, NumberOfChoice):
    b = sample(lst, k=len(lst))
    p = 0
    for j in range(0, len(b)-1):
        if b[j][0] == b[j+1][0]:
            p+=1
    point.append(p)
print('Mean P when N=26, M=2 : ', np.mean(point))
print('Standard Deviation P when N=26, M=2 : ', np.std(point))
```

1- $P(A \mid\ B) = P(A \cap B) \div P(B)$

2- $A \subset B  =>$  $P(A \cap B) = P(B)$

according 1,2  => $P(A \mid\ B) = P(A) \div P(B)$

## Question 5

```python
lst = deck_generator(M=2, N=26)
NumberOfChoice = 10000
res_Greater12 = []
res_Greater06 = []
res = []
for i in range(0, NumberOfChoice):
    b = sample(lst, k=len(lst))
    p = 0
    for j in range(0, len(b)-1):
        if b[j][0] == b[j+1][0]:
            p+=1
    if p>12:
        res_Greater12.append(1)
        res_Greater06.append(1)
    elif p>6:
        res_Greater12.append(0)
        res_Greater06.append(1)
    else:
        res_Greater12.append(0)
        res_Greater06.append(0)
    res.append(np.sum(res_Greater12) / np.sum(res_Greater06))
print('P(Point>12|Point>6) when N=26, M=2 : ',np.mean(res))
```
